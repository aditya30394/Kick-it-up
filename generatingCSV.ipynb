```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
import pandas as pd
```

```python
#Location of your data
directory = "/Users/Anurag/Desktop/DataMiningProject/Data/Kickstarter/August2019"

projectsDF = pd.DataFrame() #stores dataframes by year
pledgeAmounts = {} #Stores total amount pledged monthly per year
l = []

for filename in sorted(os.listdir(directory)):
    #print(filename)
    data = pd.read_csv(directory+'/'+filename,engine='python',error_bad_lines=False)
    l.append(data)

projectsDF = pd.concat(l)
projectsDF.head()
```

```python
########Drop Columns that aren't useful features##############
colDrop = ['converted_pledged_amount','creator','currency_symbol','currency_trailing_code','friends','fx_rate','is_backing','is_starrable','is_starred','permissions','photo','profile','source_url','usd_type','urls','static_usd_rate']
projectsDF = projectsDF.drop(colDrop, axis = 1) 
projectsDF.columns
```

```python
projectsDF['avgAmountPledged'] = (projectsDF['usd_pledged']/projectsDF['backers_count'])
projectsDF['avgAmountPledged'] = projectsDF['avgAmountPledged'].fillna(0.0)
```

```python
projectsDF.head()
```

```python
projectsDF.to_csv("cleanedCSV.csv",index=False)
```

```python
#Convert them to type categorical
newData['name_sentiment_predictions'] = pd.Categorical(newData['name_sentiment_predictions'])
dataDummies = pd.get_dummies(newData['name_sentiment_predictions'], prefix = 'name_sentiment')

#Concatenate to the original dataframe and drop the name_sentiment_predictions variable
newData = pd.concat([newData, dataDummies], axis=1)

newData['blurb_sentiment_predictions'] = pd.Categorical(newData['blurb_sentiment_predictions'])
dataDummies = pd.get_dummies(newData['blurb_sentiment_predictions'], prefix = 'blurb_sentiment')

#Concatenate to the original dataframe and drop the name_sentiment_predictions variable
newData = pd.concat([newData, dataDummies], axis=1)
```

```python
newData = newData.drop(['name_sentiment_predictions','blurb_sentiment_predictions'],axis=1)
```
